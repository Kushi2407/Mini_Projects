```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data=pd.read_csv('C:/Iris.csv')
data
```

```python
# 1. Univariate Analysis
# Visualize univariate analysis with bar plots and box plots
sns.set(style="whitegrid")
plt.figure(figsize=(12, 5))
```

## 1. Univariate Analysis

```python
# Univariate analysis for Sepal Length
plt.figure(figsize=(12, 5))
plt.subplot(1, 4, 1)
sns.histplot(data['SepalLengthCm'], kde=True, color='blue')
plt.title("Sepal Length Distribution")

# Univariate analysis for Sepal Width
plt.subplot(1, 4, 2)
sns.histplot(data['SepalWidthCm'], kde=True, color='green')
plt.title("Sepal Width Distribution")

# Univariate analysis for Petal Length
plt.subplot(1, 4, 3)
sns.histplot(data['PetalLengthCm'], kde=True, color='red')
plt.title("Petal Length Distribution")

# Univariate analysis for Petal Width
plt.subplot(1, 4, 3)
sns.histplot(data['PetalWidthCm'], kde=True, color='purple')
plt.title("Petal Width Distribution")

plt.tight_layout()
plt.show()
 

```

## 2. Bivariate Analysis

```python
# Bivariate Analysis
plt.figure(figsize=(15, 5))

# Bivariate analysis for Sepal Length vs Sepal Width
plt.subplot(1, 3, 1)
sns.scatterplot(data=data, x="SepalLengthCm", y="SepalWidthCm", hue="Species", palette="viridis")
plt.title("Sepal Length vs Sepal Width")

# Bivariate analysis for Petal Length vs Petal Width
plt.subplot(1, 3, 2)
sns.scatterplot(data=data, x="PetalLengthCm", y="PetalWidthCm", hue="Species", palette="viridis")
plt.title("Petal Length vs Petal Width")

# Bivariate analysis for Sepal Length vs Petal Length
plt.subplot(1, 3, 3)
sns.scatterplot(data=data, x="SepalLengthCm", y="PetalLengthCm", hue="Species", palette="viridis")
plt.title("Sepal Length vs Petal Length")

plt.tight_layout()
plt.show()
```

## 3. Multivariate Analysis

```python
# Multivariate Analysis (Pair Plot)
sns.set(style="whitegrid")
sns.pairplot(data, hue="Species", palette="viridis", markers=["o", "s", "D"])
plt.show()
```

## 4. Visualize above with Bar, Box, FacetGrid, Heatmap, Crosstab, Line plot


## Bar Plot

```python
# Bar Plot
plt.figure(figsize=(10, 4))
sns.countplot(data=data, x="Species", palette="viridis")
plt.title("Species Count (Bar Plot)")
```

##  Box Plot

```python
# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x="Species", y="SepalLengthCm", data=data, palette="viridis")
plt.title("Sepal Length by Species (Box Plot)")
```

## FacetGrid

```python
# FacetGrid
g = sns.FacetGrid(data, col="Species", palette="viridis")
g.map(sns.scatterplot, "SepalLengthCm", "SepalWidthCm")
g.add_legend()
plt.suptitle("FacetGrid: Sepal Length vs Sepal Width by Species", y=1.02)
```

## Heatmap

```python
# Heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu", linewidths=.5)
plt.title("Correlation Heatmap")
plt.show
```

## Crosstab Analysis

```python
# Crosstab Analysis
species_vs_sepal_length = pd.crosstab(data["Species"], data["SepalLengthCm"])
print("Crosstab Analysis - Species vs Sepal Length:")
print(species_vs_sepal_length)
```

## Line Plot

```python
# Line Plot
plt.figure(figsize=(8, 4))
sns.lineplot(x="Species", y="PetalLengthCm", data=data, palette="viridis")
plt.title("Line Plot: Petal Length by Species")
plt.show
```
